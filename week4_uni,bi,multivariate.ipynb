```python

def multivariate_eda(df):
    plt.figure(figsize=(10, 8))
    sns.pairplot(df)
    plt.title('Multivariate EDA')
    plt.show()

multivariate_eda(df[['Customer_Age', 'Avg_Utilization_Ratio', 'Credit_Limit']])
```
